```python
import boto3
import json
import aws_credentials
```

```python
#create boto3 client
client = boto3.client('kinesis', aws_access_key_id = aws_credentials.access_key_id, aws_secret_access_key = aws_credentials.aws_secret_key, region_name = 'us-east-1')
```

Create a loop for consuming the data sent through the stream to check if everything is okay;
This function is only to check if data is being sent correctly. The data stream will be collected by AWS Kinesis Firehose to be allocated on a S3 bucket.
ShardId is obtained through "put_record" method.

```python
shard = client.get_shard_iterator(StreamName = 'windfarm', ShardId = 'shardId-000000000002', ShardIteratorType = 'LATEST')['ShardIterator']
```

```python
while shard is not None:
  result = client.get_records(ShardIterator = shard)
  registry = result['Records']
  shard = result['NextShardIterator']
  for entry in registry:
    print(entry['SequenceNumber'])
    print(entry['ApproximateArrivalTimestamp'])
    print(entry['PartitionKey'])
    print(entry['Data'])
```
